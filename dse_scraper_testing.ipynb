```python
import pandas as pd
import datetime as dt
from dateutil.relativedelta import relativedelta
import bdshare
import numpy as np
pd.set_option('display.float_format', lambda x: '%.5f' % x)
```

```python
start=dt.datetime(2020,1,1)
end = dt.datetime.now()
```

```python
share_price_pivot = bdshare.get_hist_data(start,end)
```

```python
share_price_pivot.to_csv(f"share_price_pivot_{end.strftime('%d-%m-%Y')}.csv")
```

```python
share_price_pivot["close"] = share_price_pivot["close"].astype('float64')
```

```python
close_price_pivot = pd.pivot_table(share_price_pivot, values = "close", columns = "symbol", index=share_price_pivot.index)
```

```python
# turn index collumn to datetime
close_price_pivot.index = pd.to_datetime(close_price_pivot.index)
```

```python
# finds monthly return
mtl_ret = close_price_pivot.pct_change().resample('M').agg(lambda x: (x+1).prod()-1)
mtl_ret
```

```python
# returns over past 11 months 
past_11 = (mtl_ret+1).rolling(12).apply(np.prod)-1
past_11
```

```python
formation = dt.datetime(2021,1,31)
end_measurement = formation - relativedelta(months=1)
```

```python
ret_12 = past_11.loc[end_measurement]
```

```python
ret_12  = ret_12.reset_index()
```

```python
ret_12.dropna(inplace=True)
```

```python
ret_12["quintile"] = pd.qcut(ret_12.iloc[:,1], 10, duplicates='drop', labels=False)
```

```python
ret_12
```

```python
ret_12.quintile.max()
```

```python
winners = ret_12[ret_12.quintile == ret_12.quintile.max()]
losers = ret_12[ret_12.quintile==0]
```

```python
winners.columns
```

```python
winners.sort_values(by = winners.columns[1], ascending=False)
```

```python
winnerret = mtl_ret.loc[formation+relativedelta(months=1), close_price_pivot.columns.isin(winners.symbol)]
loserret = mtl_ret.loc[formation+relativedelta(months=1), close_price_pivot.columns.isin(losers.symbol)]
```

```python
x = winnerret.reset_index()
x.sort_values(by =x.columns[1], ascending=False)
```

```python
momentumprofit = winnerret.mean()
```
